```python
class hash_map:
    _hash_map_array:list = []
    def __init__(self):
        self._hash_map_array = [[] for _ in range(10)]
    def add_element(self, element:int):
        k = element % 10
        if self._hash_map_array[k]:
            self._hash_map_array[k].append(element)
        else:
            self._hash_map_array[k] = [element]
    def search_element(self, element:int):
        k = element % 10
        found = False
        return element in self._hash_map_array[k]

if __name__ == "__main__"
    obj = hash_map()
    obj.add_element(10)
    obj.add_element(11)
    print(obj.search_element(11))
    print(obj.search_element(12))            

```
